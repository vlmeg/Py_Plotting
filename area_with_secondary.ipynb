```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
from pylab import *
import seaborn
import seaborn as sns
import seaborn as sb
import pandas as pd
```

### Specifying Colors

```python
#from bokeh.palettes import Spectral11
#colors = Spectral11
#flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
#x=sns.palplot(sns.color_palette(flatui))
c1=sns.color_palette("hls", 8)
```

### Reading as df

```python
x = pd.read_csv('Workbook5.csv')
#Renaming
#x=x.rename(columns={'hydro': 'Hydro'})
x1=x[['hydro','wind',"chp_gas","chp_waste"]][1:100]
x2=x["nuclear"][1:100]
plt.figure()
sns.set_style("dark", {'axes.grid' : False})
a=x1.plot(kind="area",title="Generation", color=c1)
#a=x1.plot(kind="area",title="Generation", color=colors)
b=x2.plot(secondary_y=True, style='b')
a.set_xlabel("x-label")
a.set_ylabel("y-label")
b.set_ylabel("title")
a.legend(loc='center left', bbox_to_anchor=(1.1, 0.9))
b.legend(loc='center left', bbox_to_anchor=(1.1, 0.75))
plt.show()
```
